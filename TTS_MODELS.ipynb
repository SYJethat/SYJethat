```python
import torchaudio
import torchaudio.transforms as transforms
import numpy as np
import os
from tqdm import tqdm

DATASET_PATH = "dataset/libri/LibriSpeech/train-clean-100/"
PROCESSED_DATA_PATH = "dataset/processed/"

os.makedirs(PROCESSED_DATA_PATH, exist_ok=True)

def process_audio(file_path):
    waveform, sr = torchaudio.load(file_path)
    mel_spec_transform = transforms.MelSpectrogram(sample_rate=sr, n_mels=80)
    mel_spectrogram = mel_spec_transform(waveform)
    return mel_spectrogram.numpy()

# Process all .flac files
for subdir, _, files in os.walk(DATASET_PATH):
    for file in tqdm(files, desc="Processing Audio"):
        if file.endswith(".flac"):
            file_path = os.path.join(subdir, file)
            mel = process_audio(file_path)
            np.save(os.path.join(PROCESSED_DATA_PATH, file.replace(".flac", ".npy")), mel)

print("Preprocessing Complete!")
```

```python
import torch
import torch.nn as nn

class Tacotron2(nn.Module):
    def __init__(self, input_dim=80, hidden_dim=256, output_dim=80):
        super(Tacotron2, self).__init__()
        self.encoder = nn.LSTM(input_dim, hidden_dim, batch_first=True)
        self.decoder = nn.LSTM(hidden_dim, output_dim, batch_first=True)

    def forward(self, x):
        if x.size(-1) != 80:  # Check if the input size is correct
            raise ValueError(f"Expected input size (batch, seq_len, 80), but got {x.shape}")
        enc_out, _ = self.encoder(x)
        dec_out, _ = self.decoder(enc_out)
        return dec_out
```

```python
class HiFiGAN(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(HiFiGAN, self).__init__()
        self.conv1 = nn.Conv1d(input_dim, output_dim, kernel_size=3, padding=1)
        self.conv2 = nn.Conv1d(output_dim, output_dim, kernel_size=3, padding=1)

    def forward(self, x):
        x = torch.relu(self.conv1(x))
        x = torch.tanh(self.conv2(x))
        return x
```

```python
import torch.optim as optim

# Training Parameters
EPOCHS = 5
BATCH_SIZE = 16
LEARNING_RATE = 0.001

# Model & Optimizer
model = Tacotron2(input_dim=80, hidden_dim=256, output_dim=80)
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
loss_fn = torch.nn.MSELoss()

# Dummy DataLoader (Replace with real data)
# Corrected Shape: (batch_size, sequence_length, input_size=80)
train_data = torch.randn(100, 200, 80)  # 100 samples, 200 time-steps, 80 mel features
train_loader = torch.utils.data.DataLoader(train_data, batch_size=16, shuffle=True)


# Training Loop
for epoch in range(EPOCHS):
    for mel_spectrogram in train_loader:
        optimizer.zero_grad()
        output = model(mel_spectrogram)
        loss = loss_fn(output, mel_spectrogram)
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch+1}/{EPOCHS}, Loss: {loss.item():.4f}")

# Save model
torch.save(model.state_dict(), "tacotron2.pth")
print("Tacotron2 Training Complete!")
```

```python
# Training Parameters
EPOCHS = 5

# Model & Optimizer
model = HiFiGAN(input_dim=80, output_dim=1)
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
loss_fn = torch.nn.MSELoss()

# Training Loop
for epoch in range(EPOCHS):
    for mel_spectrogram in train_loader:
      mel_spectrogram = mel_spectrogram.permute(0, 2, 1)  # Change (B, T, C) → (B, C, T)

      optimizer.zero_grad()
      output = model(mel_spectrogram)  # Now it has correct shape
      loss = loss_fn(output, mel_spectrogram)
      loss.backward()
      optimizer.step()
    print(f"Epoch {epoch+1}/{EPOCHS}, Loss: {loss.item():.4f}")

# Save model
torch.save(model.state_dict(), "hifi_gan.pth")
print("HiFi-GAN Training Complete!")
```
